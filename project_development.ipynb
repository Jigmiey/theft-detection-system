````python
def count_upto(n):
    i=1
    while i<=n:
        yield i
        i+=1
````

````python
for j in count_upto(4):
    print(j*2)
````

````python
def my_range(start:int,stop:int,step:int):
    i = start
    while i < stop:
        yield i
    i+=step
````
